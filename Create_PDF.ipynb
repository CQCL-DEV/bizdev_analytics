```python
from __future__ import print_function
import pandas as pd
import numpy as np
df = pd.read_excel("Downloads/Data_with_Master.xlsx")
df.head()
```

```python
df_report = pd.pivot_table(df, index=["Company", "Industry", "Region"], values=["Engaged with", "Link"],
                           aggfunc=[np.sum, np.mean], fill_value=0)
df_report.head()
```

```python
from jinja2 import Environment, FileSystemLoader
env = Environment(loader=FileSystemLoader('.'))
template = env.get_template("Downloads/competitor_report_template.html")
```

```python
template_vars = {"title" : "Competitor Report - Company",
                 "national_pivot_table": df_report.to_html()}
```

```python
html_out = template.render(template_vars)
```

```python
import pdfkit

pdfkit.from_string(html_out, "competitor_report.pdf")
```
