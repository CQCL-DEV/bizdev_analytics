```python
import sqlite3

# Connect to a database or create one if it doesn't exist
conn = sqlite3.connect('example.db')

# Create a cursor object to execute SQL commands
c = conn.cursor()

# Create a table
c.execute('''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER NOT NULL,
    email TEXT UNIQUE NOT NULL
)
''')

# Insert data into the table
c.execute("INSERT INTO users (name, age, email) VALUES (?, ?, ?)", ("Alice", 30, "alice@example.com"))
c.execute("INSERT INTO users (name, age, email) VALUES (?, ?, ?)", ("Bob", 25, "bob@example.com"))
c.execute("INSERT INTO users (name, age, email) VALUES (?, ?, ?)", ("Charlie", 22, "charlie@example.com"))

# Commit the changes
conn.commit()

# Query the table
c.execute("SELECT * FROM users")
rows = c.fetchall()
for row in rows:
    print(row)

# Update a record
c.execute("UPDATE users SET age = ? WHERE name = ?", (31, "Alice"))
conn.commit()

# Delete a record
c.execute("DELETE FROM users WHERE name = ?", ("Charlie",))
conn.commit()

# Close the connection
conn.close()
```
